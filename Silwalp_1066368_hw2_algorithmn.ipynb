```python
#step1
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

# Load Iris dataset
X, y = datasets.load_iris(return_X_y=True)

# Check the size of variables X and y
X_shape = X.shape
y_shape = y.shape

# Display unique classes in y
unique_classes = set(y)
# Separate data by classes
x0 = X[y == 0]
x1 = X[y == 1]
x2 = X[y == 2]
# Plotting
plt.scatter(x0[:, 0], x0[:, 1], c='navy', label='Class 0')
plt.scatter(x1[:, 0], x1[:, 1], c='green', label='Class 1')
plt.scatter(x2[:, 0], x2[:, 1], c='orange', label='Class 2')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show(), (X_shape, y_shape), unique_classes

```

```python
#step2
from sklearn import mixture

# Create and fit the Gaussian Mixture model
gm = mixture.GaussianMixture(n_components=3)
gm.fit(X[:, 0:2])

# Predict the class of each observation
y_predict = gm.predict(X[:, 0:2])

# Separate data by predicted classes
c0 = X[y_predict == 0, 0:2]
c1 = X[y_predict == 1, 0:2]
c2 = X[y_predict == 2, 0:2]

# Plotting
plt.figure(figsize=(12, 6))

# Subplot for predicted classes
plt.subplot(1, 2, 2)
plt.scatter(c0[:, 0], c0[:, 1], c='green', label='Predicted Class 0')
plt.scatter(c1[:, 0], c1[:, 1], c='navy', label='Predicted Class 1')
plt.scatter(c2[:, 0], c2[:, 1], c='orange', label='Predicted Class 2')
plt.title("Predicted Classes")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()

# Subplot for original classes
plt.subplot(1, 2, 1)
plt.scatter(x0[:, 0], x0[:, 1], c='navy', label='Original Class 0')
plt.scatter(x1[:, 0], x1[:, 1], c='green', label='Original Class 1')
plt.scatter(x2[:, 0], x2[:, 1], c='orange', label='Original Class 2')
plt.title("Original Classes")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()

plt.show()
```
