The lifetime, T in hours of a certain brand of battery are assumed to be normally
distributed with a mean of 825 and a variance of 900.
1. Whatâ€™s the pdf (probability density function) of this distribution?


```python
# the mean and standard deviation
mean = 825  # Mean (mu)
standard_deviation = 30  # Standard deviation (sigma)

# Creating a range of x values from (mean - 3*sigma) to (mean + 3*sigma)
x_values = np.arange(mean - 3*standard_deviation, mean + 3*standard_deviation, 1)  # 1 hour increments

# Calculating the pdf values for each x
pdf_values = (1 / (standard_deviation * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_values - mean) / standard_deviation) ** 2)

# Plotting the pdf again
plt.figure(figsize=(10, 6))
plt.plot(x_values, pdf_values, label="PDF of Battery Life")
plt.xlabel("Battery Life (hours)")
plt.ylabel("Probability Density")
plt.title("Probability Density Function of Battery Lifetime")
plt.grid(True)
plt.legend()
plt.show()
```

. Find the probability that one such battery will last between 801 and 870 hours

```python
from scipy.stats import norm

# Points for which we want to find the probability
x1 = 801
x2 = 870

# Calculating z-scores for x1 and x2
z1 = (x1 - mean) / standard_deviation
z2 = (x2 - mean) / standard_deviation

# Calculating the CDF values for these z-scores
cdf1 = norm.cdf(z1)
cdf2 = norm.cdf(z2)

# The probability that the battery lasts between 801 and 870 hours
probability = cdf2 - cdf1
probability
```
